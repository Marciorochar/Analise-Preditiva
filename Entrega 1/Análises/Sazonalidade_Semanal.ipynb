#ðŸ“ˆ AnÃ¡lise de Sazonalidade Semanal nas Vendas de Camisetas BÃ¡sicas Masculinas (2022â€“2024)

Os grÃ¡ficos ilustram o comportamento das vendas de camisetas bÃ¡sicas masculinas ao longo dos dias da semana, com base em dados coletados entre janeiro de 2022 a novembro de 2024. O primeiro grÃ¡fico apresenta a soma total de vendas por dia da semana, revelando um volume maior nos fins de semana, especialmente aos sÃ¡bados, sugerindo uma preferÃªncia do consumidor por compras nesse perÃ­odo. JÃ¡ o segundo grÃ¡fico exibe a mÃ©dia diÃ¡ria de vendas por dia da semana, que segue um padrÃ£o semelhante: os fins de semana continuam em destaque, com mÃ©dias consistentemente mais altas em relaÃ§Ã£o aos dias Ãºteis. Esses resultados indicam uma sazonalidade semanal clara, com picos nos sÃ¡bados e domingos, o que pode estar associado a maior disponibilidade de tempo para compras, seja em lojas fÃ­sicas ou online, e a campanhas promocionais pontuais.



```python
# Instalar bibliotecas necessÃ¡rias
!pip install pandas matplotlib openpyxl

# Imports
import pandas as pd
import matplotlib.pyplot as plt

# Upload do arquivo Excel
from google.colab import files
uploaded = files.upload()

# Ler o arquivo Excel enviado
nome_arquivo = list(uploaded.keys())[0]
df = pd.read_excel(nome_arquivo)

# Converter colunas
df['Timestamp'] = pd.to_datetime(df['Timestamp'], dayfirst=True, errors='coerce')
df['Camisetas_bÃ¡sicas_masculinas'] = pd.to_numeric(df['Camisetas_bÃ¡sicas_masculinas'], errors='coerce')

# Criar coluna com dia da semana em inglÃªs
df['dia da semana'] = df['Timestamp'].dt.day_name()

# Traduzir dias para portuguÃªs manualmente
dias_pt = {
    'Monday': 'Segunda',
    'Tuesday': 'TerÃ§a',
    'Wednesday': 'Quarta',
    'Thursday': 'Quinta',
    'Friday': 'Sexta',
    'Saturday': 'SÃ¡bado',
    'Sunday': 'Domingo'
}
df['dia da semana'] = df['dia da semana'].map(dias_pt)

# Agrupar por dia da semana (na ordem correta)
ordem_dias = ['Segunda', 'TerÃ§a', 'Quarta', 'Quinta', 'Sexta', 'SÃ¡bado', 'Domingo']
vendas_semanais = df.groupby('dia da semana')['Camisetas_bÃ¡sicas_masculinas'].sum().reindex(ordem_dias)

# Plotar grÃ¡fico
plt.figure(figsize=(10, 6))
plt.bar(vendas_semanais.index, vendas_semanais.values, color='mediumseagreen')
plt.title('Sazonalidade Semanal - Vendas de Camisetas BÃ¡sicas Masculinas')
plt.xlabel('Dia da Semana')
plt.ylabel('Quantidade de Vendas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# ðŸ“¦ Instalar bibliotecas necessÃ¡rias (caso ainda nÃ£o estejam instaladas)
!pip install matplotlib openpyxl

# ðŸ“š Imports
import pandas as pd
import matplotlib.pyplot as plt

# ðŸ“¤ Upload do arquivo Excel (executa no Colab)
from google.colab import files
uploaded = files.upload()

# ðŸ“„ LÃª o arquivo Excel enviado
arquivo_enviado = list(uploaded.keys())[0]
df = pd.read_excel(arquivo_enviado)

# ðŸ•“ Converte a coluna de datas
df['Timestamp'] = pd.to_datetime(df['Timestamp'], dayfirst=True, errors='coerce')

# ðŸ”¢ Converte a coluna de vendas para nÃºmero (tratando erros)
df['Camisetas_bÃ¡sicas_masculinas'] = pd.to_numeric(df['Camisetas_bÃ¡sicas_masculinas'], errors='coerce')

# ðŸ—“ï¸ Cria uma nova coluna com o nome do dia da semana em inglÃªs
df['Dia_Semana_EN'] = df['Timestamp'].dt.day_name()

# ðŸ—ºï¸ Traduz os dias para portuguÃªs
dias_pt = {
    'Monday': 'Segunda',
    'Tuesday': 'TerÃ§a',
    'Wednesday': 'Quarta',
    'Thursday': 'Quinta',
    'Friday': 'Sexta',
    'Saturday': 'SÃ¡bado',
    'Sunday': 'Domingo'
}
df['Dia_da_Semana'] = df['Dia_Semana_EN'].map(dias_pt)

# ðŸ“Š Agrupa e calcula a mÃ©dia de vendas por dia da semana
media_vendas = df.groupby('Dia_da_Semana')['Camisetas_bÃ¡sicas_masculinas'].mean().reindex(
    ['Segunda', 'TerÃ§a', 'Quarta', 'Quinta', 'Sexta', 'SÃ¡bado', 'Domingo']
)

# ðŸ“ˆ Plota o grÃ¡fico
plt.figure(figsize=(10, 6))
plt.bar(media_vendas.index, media_vendas.values, color='mediumseagreen')
plt.title('MÃ©dia de Vendas por Dia da Semana - Camisetas BÃ¡sicas Masculinas')
plt.xlabel('Dia da Semana')
plt.ylabel('MÃ©dia de Vendas')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
